```python
import pandas as pd
```

```python
# Read map_bank_coa.csv and read all field as string
df = pd.read_csv('map_bank_coa.csv', dtype=str)
df.tail(10)
```

### Fill e-wallet

```python
# Fill MapBank_EDC_CoaID, MapBank_EDC_CoaAccountNo, MapBank_EDC_CoaDescription when its still null by looking up the previous row with the same MapBank_NatBankID, MapBank_NatBankIsEDC, where MapBank_BankAccountNo == 'EDC' and MapBank_IsActive == Y
df['MapBank_EDC_CoaID'] = df.loc[(df['MapBank_NatBankIsEDC'] == 'Y') & (df['MapBank_BankAccountNo'] == 'EDC') ].groupby(['MapBank_NatBankID'])['MapBank_EDC_CoaID'].ffill()
df['MapBank_EDC_CoaAccountNo'] = df.loc[(df['MapBank_NatBankIsEDC'] == 'Y') & (df['MapBank_BankAccountNo'] == 'EDC') ].groupby(['MapBank_NatBankID'])['MapBank_EDC_CoaAccountNo'].ffill()
df['MapBank_EDC_CoaDescription'] = df.loc[(df['MapBank_NatBankIsEDC'] == 'Y') & (df['MapBank_BankAccountNo'] == 'EDC') ].groupby(['MapBank_NatBankID'])['MapBank_EDC_CoaDescription'].ffill()
df.tail(10)
```

## Output CSV

```python
# Convert cell that NaN to empty string
df['MapBank_EDC_CoaID'] = df['MapBank_EDC_CoaID'].fillna('')
df['MapBank_EDC_CoaAccountNo'] = df['MapBank_EDC_CoaAccountNo'].fillna('')
df['MapBank_EDC_CoaDescription'] = df['MapBank_EDC_CoaDescription'].fillna('')
df['MapBank_EDC_CoaID'] = df['MapBank_EDC_CoaID'].fillna('')
df['MapBank_EDC_CoaAccountNo'] = df['MapBank_EDC_CoaAccountNo'].fillna('')
df['MapBank_EDC_CoaDescription'] = df['MapBank_EDC_CoaDescription'].fillna('')

# save csv
df.to_csv('map_bank_coa_out.csv', index=0)

df = pd.read_csv('map_bank_coa_out.csv')
df.tail(10)
```

## Convert CSV into SQL INSERT INTO

```python
# Convert df nan to null
df = df.where(pd.notnull(df), None)

# Convert CSV into SQL INSERT INTO
with open('map_bank_coa.sql', 'w') as f:
    f.write ("INSERT INTO `map_bank_coa` (`MapBank_ID`, `MapBank_BranchCode`, `MapBank_NatBankID`, `MapBank_NatBankCode`, `MapBank_NatBankName`, `MapBank_NatBankIsEDC`, `MapBank_BankAccountNo`, `MapBank_CoaID`, `MapBank_CoaAccountNo`, `MapBank_CoaDescription`, `MapBank_EDC_CoaID`, `MapBank_EDC_CoaAccountNo`, `MapBank_EDC_CoaDescription`) VALUES \n")
    for i in range(len(df)):
        f.write(f"('{df.iloc[i, 0]}', '{df.iloc[i, 1]}', '{df.iloc[i, 2]}', '{df.iloc[i, 3]}', '{df.iloc[i, 4]}', '{df.iloc[i, 5]}', '{df.iloc[i, 6]}', '{df.iloc[i, 7]}', '{df.iloc[i, 8]}', '{df.iloc[i, 9]}', '{df.iloc[i, 10]}', '{df.iloc[i, 11]}', '{df.iloc[i, 12]}'),\n")

    f.write ("ON DUPLICATE KEY UPDATE `MapBank_ID` = VALUES(`MapBank_ID`), `MapBank_BranchCode` = VALUES(`MapBank_BranchCode`), `MapBank_NatBankID` = VALUES(`MapBank_NatBankID`), `MapBank_NatBankCode` = VALUES(`MapBank_NatBankCode`), `MapBank_NatBankName` = VALUES(`MapBank_NatBankName`), `MapBank_NatBankIsEDC` = VALUES(`MapBank_NatBankIsEDC`), `MapBank_BankAccountNo` = VALUES(`MapBank_BankAccountNo`), `MapBank_CoaID` = VALUES(`MapBank_CoaID`), `MapBank_CoaAccountNo` = VALUES(`MapBank_CoaAccountNo`), `MapBank_CoaDescription` = VALUES(`MapBank_CoaDescription`), `MapBank_EDC_CoaID` = VALUES(`MapBank_EDC_CoaID`), `MapBank_EDC_CoaAccountNo` = VALUES(`MapBank_EDC_CoaAccountNo`), `MapBank_EDC_CoaDescription` = VALUES(`MapBank_EDC_CoaDescription`");
```
