### Depok

```python
df = pd.read_csv(f"{FILE_PATH}/regpas_diagnosa_depok.csv")
df.head()
```

```python
# Explore the data through dataframe
df.info()
```

```python
# Count unique values, null values, and data types
df.nunique()
```

```python
# Explore most frequent values
df.describe()
```

```python
# Display top-10 most frequent values for each columns
for col in df.columns:
    print(f"Top-10 most frequent values for column {col}")
    print(df[col].value_counts().head(10))
    print("\n")

# Create visualization for each columns
import matplotlib.pyplot as plt
# Create visualization for each column
plt.figure(figsize=(10, 6))

# Top-10 most frequent values for column DiagnosaUtama
plt.subplot(1, 2, 1)
df['DiagnosaUtama'].value_counts().head(10).plot(kind='bar')
plt.title('Top-10 most frequent values for column DiagnosaUtama')
plt.xlabel('DiagnosaUtama')
plt.ylabel('Count')

# Top-10 most frequent values for column DiagnosaKerja
plt.subplot(1, 2, 2)
df['DiagnosaKerja'].value_counts().head(10).plot(kind='bar')
plt.title('Top-10 most frequent values for column DiagnosaKerja')
plt.xlabel('DiagnosaKerja')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Search row when 'DiagnosaUtama' = J00 Group by DiagnosaKerja and add count the frequent for each DiagnosaKerja
df[df['DiagnosaUtama'] == 'J00']['DiagnosaKerja'].value_counts().head(15).plot(kind='bar')
plt.title('Top 15 Ragam DiagnosaKerja untuk ICD10 = J00')
plt.xlabel('')
plt.ylabel('Jumlah')

# Add legend
total_data = len(df[df['DiagnosaUtama'] == 'J00'])
total_variation = len(df[df['DiagnosaUtama'] == 'J00']['DiagnosaKerja'].value_counts())
plt.legend(
    [f'Total Data: {total_data}\nTotal Variasi DiagnosaKerja: {total_variation}'],
    loc='upper right',
    frameon=True
)
print(f"Total Data: {total_data}")
print(f"Total Variation: {total_variation}")

plt.show()
```

```python
# Search row when 'DiagnosaUtama' = J00 Group by DiagnosaKerja and add count the frequent for each DiagnosaKerja
df[df['DiagnosaKerja'] == 'imunisasi']['DiagnosaUtama'].value_counts().plot(kind='bar')
plt.title('Ragam ICD-10 untuk Diagnosa = "imunisasi"')
plt.xlabel('')
plt.ylabel('Jumlah')
plt.legend(
    [f'Total Data: {total_data}\nTotal Variasi ICD10: {total_variation}'],
    loc='upper right',
    frameon=True
)
```
