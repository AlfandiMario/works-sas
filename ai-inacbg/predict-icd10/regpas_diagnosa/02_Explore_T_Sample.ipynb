```python
import pandas as pd
import matplotlib.pyplot as plt

FILE_PATH = '~/ai-inacbg/predict-icd10/t_sample.csv'
df = pd.read_csv(FILE_PATH)
# Remove field T_SampleFileName, T_SampleIsActive, T_SampleCreated, T_SampleLastUpdated
df = df.drop(columns=['T_SampleID','T_SampleFileName', 'T_SampleIsActive', 'T_SampleCreated', 'T_SampleLastUpdated'])
# Change column T_SampleCode become Code
df = df.rename(columns={'T_SampleCode': 'ICD10'})
df = df.rename(columns={'T_SampleName': 'Diagnose'})
```

```python
# Display top-10 most frequent values for each columns
for col in df.columns:
    print(f"Top-10 most frequent values for column {col}")
    print(df[col].value_counts().head(10))
    print("\n")

# Create visualization for each columns
import matplotlib.pyplot as plt
# Create visualization for each column
plt.figure(figsize=(10, 6))

# Top-10 most frequent values for column DiagnosaUtama
plt.subplot(1, 2, 1)
df['ICD10'].value_counts().head(10).plot(kind='bar')
plt.title('Top-10 most frequent values for column ICD10')
plt.xlabel('ICD10')
plt.ylabel('Count')

# Top-10 most frequent values for column DiagnosaKerja
plt.subplot(1, 2, 2)
df['Diagnose'].value_counts().head(10).plot(kind='bar')
plt.title('Top-10 most frequent values for column Diagnose')
plt.xlabel('Diagnose')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Display top-100 most frequent values for each columns
top100_icd10 = df['ICD10'].value_counts().head(100)
top100_diagnose = df['Diagnose'].value_counts().head(101) # Pertama, diagnose = -

print(f"Top-100 most frequent ICD10 values")
for i, (icd10, count) in enumerate(top100_icd10.items(), 1):
    print(f"{i}. {icd10} \t({count} rows)")

print("\n")
print(f"Top-100 most frequent Diagnose values")
for i, (diagnose, count) in enumerate(top100_diagnose.items(), 1):
    print(f"{i}. {diagnose} \t({count} rows)")
```

```python
# Create a horizonal bar chart for top-100 most frequent values for column ICD10
plt.figure(figsize=(10, 20))
top100_icd10.plot(kind='barh')
plt.title('Top-100 most frequent values for column ICD10')
plt.xlabel('Count')
plt.ylabel('ICD10')
plt.gca().invert_yaxis()
plt.grid(True)  # Add grid
plt.show()
```

```python
# Create a horizonal bar chart for top-100 most frequent values for column Diagnose
plt.figure(figsize=(10, 20))
top100_diagnose.plot(kind='barh')
plt.title('Top-100 most frequent values for column Diagnose')
plt.xlabel('Count')
plt.ylabel('Diagnose')
plt.gca().invert_yaxis()
plt.grid(True)  # Add grid
plt.show()
```
