```python
import pandas as pd
icd10 = pd.read_csv('ICD10codes.csv', header=None)
```

```python
def add_dot(s):
    return s[:3]+'.'+s[3:]

for i in range(len(icd10)):
    if len(icd10[0].iloc[i])>3:
        icd10.loc[i, 4] = icd10[4].iloc[i] + ', ' + icd10[5].iloc[i]
        icd10.loc[i, 2] = add_dot(icd10[2].iloc[i])

icd10 = icd10.drop([0,1,3,5], axis=1)
icd10 = icd10.set_axis(['icd10_code', 'diagnoses'], axis=1)

for i in range(len(icd10)):
    if '.' not in icd10['icd10_code'].iloc[i]:
        icd10.loc[i, 'icd10_code'] = add_dot(icd10['icd10_code'].iloc[i])
```

```python
icd10.to_csv('labeled_icd10_flanT5.csv', index=False)
labeled_icd10 = pd.read_csv('labeled_icd10_flanT5.csv', header=0)
```

```python
import json

with open('labeled_icd10_flanT5.jsonl', 'w') as f:
    for _, row in df.iterrows():
        entry = {
            "prompt": f"Diagnosis: {row['diagnoses']}\nICD-10-CM Code:",
            "completion": row['icd10_code']
        }
        f.write(json.dumps(entry) + '\n')
```
